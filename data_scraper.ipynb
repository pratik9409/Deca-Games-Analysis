```python
import pandas as pd
from google_play_scraper import Sort, reviews

# Fetch the initial set of reviews
review_result, continuation_token = reviews(
    'fi.twomenandadog.zombiecatchers',
    lang='en',
    country='us',
    sort=Sort.NEWEST,
    count=7
)

# Fetch additional reviews using the continuation token
more_result, _ = reviews(
    'fi.twomenandadog.zombiecatchers',
    continuation_token=continuation_token
)

# Combine the results
all_reviews = review_result + more_result

# Ensure all datetime objects are converted to strings for CSV export
for review in all_reviews:
    review['at'] = review['at'].strftime('%Y-%m-%d %H:%M:%S') if review['at'] else None
    review['repliedAt'] = review['repliedAt'].strftime('%Y-%m-%d %H:%M:%S') if review['repliedAt'] else None

# Convert the list of reviews into a DataFrame
review_df = pd.DataFrame(all_reviews)

# Save DataFrame to a CSV file
review_df.to_csv('zombie_catchers_reviews.csv', index=False)

# Display the DataFrame to verify
print(review_df)
```

```python
import pandas as pd
from google_play_scraper import Sort, reviews

# List of app IDs
app_ids = [
    'com.iugome.lilknights',
    'fi.twomenandadog.zombiecatchers',
    'com.backflipstudios.android.dragonvale',
    'com.beesquare.almostahero',
    'com.fridaysgames.godsandglory',
    'com.chillingo.robberybobfree.android.row',
    'com.chillingo.robberybob2.android.gplay',
    'com.kabam.evd',
    'com.kabam.doamobile',
    'com.kabam.newcnhoc',
    'com.kabam.kocmobile',
    'com.kabam.underworldandroid',
    'com.sixwaves.cml',
    'com.empirestudiosinc.cmg',
    'com.onethumbmobile.celticheroes',
    'com.madfingergames.unkilled',
    'com.madfingergames.SamuraiIIAll',
    'com.madfingergames.deadtrigger',
    'com.madfingergames.deadtrigger2',
    'com.madfingergames.legends',
    'com.nextgames.android.twd'
]

# List to store all reviews
all_reviews = []

# Loop through the app IDs and fetch reviews
for app_id in app_ids:
    # Fetch the initial set of reviews
    review_result, continuation_token = reviews(
        app_id,
        lang='en',
        country='us',
        sort=Sort.NEWEST,
        count=7
    )
    
    # Fetch additional reviews using the continuation token
    more_result, _ = reviews(
        app_id,
        continuation_token=continuation_token
    )
    
    # Combine the results
    reviews_combined = review_result + more_result
    
    # Ensure all datetime objects are converted to strings for CSV export
    for review in reviews_combined:
        review['appId'] = app_id  # Add appId to each review
        review['at'] = review['at'].strftime('%Y-%m-%d %H:%M:%S') if review['at'] else None
        review['repliedAt'] = review['repliedAt'].strftime('%Y-%m-%d %H:%M:%S') if review['repliedAt'] else None
    
    # Append the reviews to the list
    all_reviews.extend(reviews_combined)

# Convert the list of reviews into a DataFrame
review_df = pd.DataFrame(all_reviews)

# Save DataFrame to a CSV file
review_df.to_csv('google_play_store_reviews.csv', index=False)

# Display the DataFrame to verify
print(review_df)
```

```python
import pandas as pd
from app_store_scraper import AppStore
from datetime import datetime

# List of app IDs and names
app_ids = [
    {'name': 'Castle Age HD', 'id': 475561257},
    {'name': 'Call me a Gangsta', 'id': 1530493373},
    {'name': 'Call me a Legend', 'id': 1449121741},
    {'name': 'Shadowgun Legends: Online FPS', 'id': 1091251242},
    {'name': 'Elves vs. Dwarves', 'id': 1231491435},
    {'name': 'Almost a Hero', 'id': 1116630619},
    {'name': 'Gods and Glory', 'id': 1050101469},
    {'name': 'Unkilled', 'id': 969488951},
    {'name': 'Dead Trigger 2', 'id': 720063540},
    {'name': 'Heroes of Camelot', 'id': 570964996},
    {'name': 'Dragons of Atlantis', 'id': 558756671},
    {'name': 'Underworld Empire', 'id': 555150599},
    {'name': 'Dead Trigger', 'id': 533079551},
    {'name': 'Kingdoms of Camelot: Battle', 'id': 476546099},
    {'name': 'Celtic Heroes', 'id': 431837446},
    {'name': 'Samurai 2: Vengeance', 'id': 392486160}
]

# Lists to store app data and reviews
all_app_data = []
all_reviews = []

# Loop through the app IDs and fetch details and reviews
for app in app_ids:
    app_store = AppStore(country="us", app_name=app['name'], app_id=app['id'])
    
    # Fetch reviews (up to 100 reviews)
    app_store.review(how_many=10)
    
    # Store app details
    app_data = {
        'appId': app['id'],
        'appName': app['name'],
        'url': app_store.url,
        'description': app_store.search_id,
        # 'developer': app_store.developer,
        # 'developer_url': app_store.developer_url,
        # 'rating': app_store.rating,
        # 'ratings_count': app_store.ratings_count,
        # 'price': app_store.price,
        # 'currency': app_store.currency,
        # 'version': app_store.version,
        # 'release_date': app_store.release_date.strftime('%Y-%m-%d') if app_store.release_date else None,
        # 'last_update': app_store.updated.strftime('%Y-%m-%d') if app_store.updated else None,
        # 'size': app_store.size,
        # 'genre': app_store.genre,
        # 'content_rating': app_store.content_rating,
        # 'supported_languages': app_store.languages
    }
    all_app_data.append(app_data)
    
    # Store reviews
    for review in app_store.reviews:
        review_data = {
            'appId': app['id'],
            'appName': app['name'],
            # 'reviewId': review['id'],
            'userName': review['userName'],
            # 'userImage': review['userImage'],
            'rating': review['rating'],
            'title': review['title'],
            'content': review['review'],
            'date': review['date'].strftime('%Y-%m-%d %H:%M:%S') if review['date'] else None,
            'isEdited': review['isEdited'],
            # 'developerResponse': review['developerResponse'],
            # 'developerResponseDate': review['developerResponseDate'].strftime('%Y-%m-%d %H:%M:%S') if review['developerResponseDate'] else None
        }
        all_reviews.append(review_data)

# Convert the lists to DataFrames
app_df = pd.DataFrame(all_app_data)
reviews_df = pd.DataFrame(all_reviews)

# Save DataFrames to CSV files
app_df.to_csv('app_store_app_data.csv', index=False)
reviews_df.to_csv('app_store_reviews.csv', index=False)

# Display the DataFrames to verify
print(app_df)
print(reviews_df)
```
