```python
import pandas as pd
```

```python
# Load the CSV files into DataFrames
app_data = pd.read_csv('google_play_store_app_data.csv')
reviews_data = pd.read_csv('google_play_store_reviews.csv')
```

```python
app_data.isnull().sum()
```

```python
# Dropping the as it has no values
app_data.drop(['saleTime', 'originalPrice', 'saleText'], axis=1, inplace=True)
```

```python
# Fill missing values with appropriate placeholders
app_data['inAppProductPrice'].fillna(0, inplace=True)
app_data['video'].fillna('No video', inplace=True)
app_data['videoImage'].fillna('No video', inplace=True)
app_data['contentRatingDescription'].fillna('No description', inplace=True)
```

```python
app_data.info()
```

```python
# Convert data types
app_data['score'] = pd.to_numeric(app_data['score'], errors='coerce')
app_data['ratings'] = pd.to_numeric(app_data['ratings'], errors='coerce')
app_data['reviews'] = pd.to_numeric(app_data['reviews'], errors='coerce')
app_data['installs'] = app_data['installs'].str.replace(',', '').str.replace('+', '').astype(int)
app_data['price'] = pd.to_numeric(app_data['price'], errors='coerce')
app_data['inAppProductPrice'] = pd.to_numeric(app_data['inAppProductPrice'].str.replace('$', ''), errors='coerce')
```

```python
# Convert date columns to datetime
app_data['released'] = pd.to_datetime(app_data['released'], errors='coerce')
app_data['lastUpdatedOn'] = pd.to_datetime(app_data['lastUpdatedOn'], errors='coerce')
```

```python
app_data.columns
```

```python
# Replace True/False with 1/0 so it will be easy while importing to sql
boolean_columns = ['free', 'sale', 'offersIAP', 'adSupported', 'containsAds']

# Check if the columns exist in the DataFrame
for col in boolean_columns:
    if col in app_data.columns:
        # Convert booleans or string representations of booleans to integers
        app_data[col] = app_data[col].map({True: 1, False: 0, 'True': 1, 'False': 0})
```

```python
app_data['free']
```

```python
reviews_data.isnull().sum()
```

1) reviewCreatedVersion: Missing values here might mean the review was created before the versioning was tracked. We can fill these with a placeholder like "Unknown".

2) replyContent and repliedAt: If the review hasn't been replied to, these can remain null or be filled with a placeholder like "No reply".

3) appVersion: Similar to reviewCreatedVersion, fill missing values with "Unknown".

```python
# Fill missing values
reviews_data['reviewCreatedVersion'].fillna('Unknown', inplace=True)
reviews_data['replyContent'].fillna('No reply', inplace=True)
reviews_data['repliedAt'].fillna('No reply', inplace=True)
reviews_data['appVersion'].fillna('Unknown', inplace=True)
```

```python
reviews_data.info()
```

```python
 # Convert date columns to datetime
reviews_data['at'] = pd.to_datetime(reviews_data['at'], errors='coerce')
```

```python
import chardet

file_path = 'D:\decagames\google_play_store_reviews_clean.csv'
# Read the first few bytes to detect the encoding
with open(file_path, 'rb') as file:
    raw_data = file.read(10000)  # Read the first 10000 bytes
    result = chardet.detect(raw_data)
    detected_encoding = result['encoding']
    
detected_encoding
```

```python
# Read the file with detected encoding and write to a new UTF-8 encoded file
input_file_path = 'google_play_store_reviews_clean.csv'
output_file_path = 'google_play_store_reviews_clean_utf8.csv'

with open(input_file_path, 'r', encoding=detected_encoding, errors='ignore') as infile:
    with open(output_file_path, 'w', encoding='utf-8') as outfile:
        for line in infile:
            outfile.write(line)
```
